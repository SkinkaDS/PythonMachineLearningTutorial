```python
# Loadding libs
import pandas
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import numpy as np
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.preprocessing import OneHotEncoder

import warnings
warnings.filterwarnings("ignore")
```

```python
# Usando o Pandas para importar os dados importando os dados, no caso, algumas partidas de lol que achei no kaggle
url = "data/games.csv"
dataset = pandas.read_csv(url)
```

```python
# A função shape nos retorna as dimenções do dataset - No nosso caso, 51490 observações de 61 Variáveis
dataset.shape
```

```python
# A função head nos retorna a cabeça de um dataset, a quantidade é passada como parametro
dataset.head(10)
```

```python
# Histograma da duração das partidas, só pra ilustrar como plotar gráficos

dataset.gameDuration.hist()
plt.show()
```

```python
# Aqui transformamos nosso conjunto de dados em um array
array = dataset.values

# Aqui removemos do conjunto a variavel que queremos prever, o vencedor, q está na posição 4 do array
# Usamos a função delete do numpy para isso
X = np.delete(array,4,1)

# Aqui usamos o split do python pra pegar apenas a coluna que queremos prever
Y = array[:,4]

# Defininmos o tamanho do conjunto de validação
validation_size = 0.20

# Apenas um número aleatorio qualquer, nao importa muito qual
seed = 7

# Finalmente dividindo os dados usando uma função sklearn
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

```python
# Usando a acuracia como métrica de avaliação do mod
scoring = 'accuracy'

# Usando alguns dos modelos principais no aprendizado de maquina
models = []
models.append(('LR', LogisticRegression())) 
models.append(('LDA', LinearDiscriminantAnalysis())) 
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
# models.append(('SVM', SVC())) #Esse aqui demora, se tiver com paciencia, descomente essa linha
```

```python
# Aqui basicamente iteramos modelo a modelo e verificamos a acurácia de cada um deles, usamos algumas funções do sklearn
# vale a pena pesquisar um pouco sobre o que faz cada uma das função na propria documentação do sklearn
# https://scikit-learn.org/stable/documentation.html
results = []
names = []
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
# Apesar de ter ficado bem esparço, esse é um gráfico famoso na estatistica, chamado de box plot, Diagrama de Extremos e Quartis,
# ou, para os intimos, gráfico maria chiquinha, vale a pena pesquisar como interpretar

fig = plt.figure()
fig.suptitle('Comparação dos Modelos')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
